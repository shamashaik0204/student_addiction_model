# Linear Regression(Bias and variance values)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from mlxtend.evaluate import bias_variance_decomp
df = pd.read_csv(r"C:\Users\Shakil\OneDrive\Desktop\Project\Students Social Media Addiction.csv")
obj = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = obj.fit_transform(df[col])
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = LinearRegression()
# ---------- bias variance ----------
mse, bias, var = bias_variance_decomp(
    model,
    X_train, y_train,
    X_test, y_test,
    loss='mse',
    num_rounds=200,
    random_seed=1
)
print('MSE:', mse)
print('Bias:', bias)
print('Variance:', var)
```

# PipeLine Creation:

```python
print("Step 1: Import libraries")
import time
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.pipeline import Pipeline as IMBPipeline

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
from imblearn.over_sampling import SMOTE

from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, f1_score

from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB


print("Step 2: Load dataset")
df = pd.read_csv(r"C:\Users\Shakil\OneDrive\Desktop\Project\Students Social Media Addiction.csv")

# remove rare classes
counts = df['Addicted_Score'].value_counts()
df = df[df['Addicted_Score'].isin(counts[counts >= 5].index)]

X = df.drop(columns=['Addicted_Score'])
y = df['Addicted_Score']


print("Step 4: Train Test Split (STRATIFIED)")
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)


print("Step 5: Creating pipeline")
def model_pipeline(X, model):

    num_cols = X.select_dtypes(include=np.number).columns
    cat_cols = X.select_dtypes(exclude=np.number).columns

    numeric_pipeline = Pipeline([
        ('scaler', MinMaxScaler())
    ])

    categorical_pipeline = Pipeline([
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, num_cols),
        ('cat', categorical_pipeline, cat_cols)
    ])

    pipe = IMBPipeline([
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42)),
        ('feature_selection', SelectKBest(chi2, k=10)),
        ('model', model)
    ])

    return pipe


print("Step 6: Model selection function")

def select_model(X, y):

    models = {
        "RandomForest": RandomForestClassifier(),
        "DecisionTree": DecisionTreeClassifier(),
        "KNN": KNeighborsClassifier(),
        "Ridge": RidgeClassifier(),
        "BernoulliNB": BernoulliNB(),
        "SVC": SVC(probability=True)
    }

    results = []

    for name, clf in models.items():

        print("Training:", name)
        start = time.time()

        pipe = model_pipeline(X, clf)

        scores = cross_val_score(
            pipe,
            X,
            y,
            cv=5,
            scoring='roc_auc_ovr'
        )

        results.append({
            "Model": name,
            "ROC_AUC": scores.mean(),
            "Time(min)": round((time.time()-start)/60,2)
        })

    return pd.DataFrame(results).sort_values(by="ROC_AUC", ascending=False)


print("Step 7: Comparing models")
model_results = select_model(X_train, y_train)
print(model_results)


print("Step 8: Train best model (SVC example)")
final_model = SVC(probability=True)
final_pipeline = model_pipeline(X_train, final_model)

final_pipeline.fit(X_train, y_train)


print("Step 9: Prediction")
y_pred = final_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))


print("\nStep 10: ROC AUC")
if hasattr(final_pipeline.named_steps['model'], "predict_proba"):
    probs = final_pipeline.predict_proba(X_test)
    roc = roc_auc_score(
        y_test,
        probs,
        multi_class='ovr',
        labels=final_pipeline.classes_
    )
else:
    roc = roc_auc_score(y_test, y_pred)

print("ROC AUC:", roc)


print("\nStep 11: Classification Report\n")
print(classification_report(y_test, y_pred))
```

```python
print("\nStep 12: Saving model")

import pickle

with open("student_addiction_model.pkl", "wb") as f:
    pickle.dump(final_pipeline, f)

print("Model saved successfully!")
```
